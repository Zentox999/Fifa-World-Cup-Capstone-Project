```python
#Bar Plot [use sns for 1 instead of 2 graphs]
team1 = fifa['team1'].value_counts().nlargest(5)
team1.plot(kind='bar', color=['green'])
plt.title('Country Goals')
plt.xlabel('Teams')
plt.ylabel('Goals')
plt.xticks(rotation=90)
plt.show()
```

```python
#Bar Plot
team2 = fifa['team2'].value_counts().nlargest(5)
team2.plot(kind='bar', color=['darkblue'])
plt.title('Country Goals')
plt.xlabel('Teams')
plt.ylabel('Goals')
plt.xticks(rotation=90)
plt.show()
```

```python
#use sns again

plt.figure(figsize = (14,6))

sns.countplot(x='team1', order=fifa["team1"].value_counts().index, data=fifa, palette = "pastel")
plt.xlabel("Team 1")
plt.ylabel("Number of Goals")
plt.title("Penalty Goals wise Count")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (14,6))

sns.countplot(x='team2', order=fifa["team2"].value_counts().index, data=fifa, palette = "Set3")
plt.xlabel("Team 2")
plt.ylabel("Number of Goals")
plt.title("Goals wise Count")
plt.xticks(rotation=90)
plt.show()
```

```python
#violin plot for team 1
plt.figure(figsize=(14,8),dpi=150)
sns.violinplot(x="team1", y="free kicks team1", data=fifa)
plt.xticks(rotation=90)
plt.show()
```

```python
#violin plot for team 2
plt.figure(figsize=(14,8))
sns.violinplot(x="team2", y="free kicks team2", data=fifa)
plt.xticks(rotation=90)
plt.show()
```
